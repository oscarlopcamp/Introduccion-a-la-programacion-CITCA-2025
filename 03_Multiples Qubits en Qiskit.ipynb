## üß© Secci√≥n 1 ‚Äì Creaci√≥n de sistemas con m√∫ltiples qubits

Hasta ahora hemos trabajado con circuitos de **un solo qubit**, donde el estado era una combinaci√≥n de `|0‚ü©` y `|1‚ü©`.

Ahora construiremos circuitos con **dos qubits**, lo cual da lugar a **cuatro posibles estados base**:

- `|00‚ü©`
- `|01‚ü©`
- `|10‚ü©`
- `|11‚ü©`

Estos forman la **base computacional** para sistemas de 2 qubits.  
Cualquier combinaci√≥n lineal de estos estados es v√°lida para describir un sistema cu√°ntico de dos qubits.

---

En Qiskit, se crea un sistema de 2 qubits as√≠:



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from IPython.display import display

# Crear circuito con 2 qubits
qc = QuantumCircuit(2)
psi = Statevector(qc)

# Mostrar circuito vac√≠o
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema |00‚ü©:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))
````

---

### ‚úÖ Explicaci√≥n del resultado

- El sistema comienza en el estado `|00‚ü©`, lo cual significa que **ambos qubits est√°n en `|0‚ü©`**.
- El vector de estado tiene **cuatro componentes**, y solo la correspondiente a `|00‚ü©` tiene valor `1`; las dem√°s son `0`.
- El orden en Qiskit sigue la convenci√≥n **little-endian**, lo que significa que el qubit **de mayor √≠ndice (qubit 1)** es el m√°s significativo.

Esto es importante para interpretar correctamente las posiciones en el vector de estado:

$$
[|00‚ü©,\ |01‚ü©,\ |10‚ü©,\ |11‚ü©]
$$

---

üìå En la siguiente secci√≥n aplicaremos compuertas a uno de los qubits para cambiar el estado, y observaremos c√≥mo se modifica el sistema completo.

# ‚öôÔ∏è Secci√≥n 2 ‚Äì Aplicaci√≥n de compuertas individuales en sistemas de 2 qubits

Ahora que hemos creado un sistema con dos qubits en estado `|00‚ü©`, aplicaremos una **compuerta X al qubit 1**, para transformar el estado a `|01‚ü©`.

Tambi√©n utilizaremos la funci√≥n `plot_state_qsphere()` para visualizar el estado resultante en una **QSphere**, una herramienta gr√°fica que representa el **estado completo del sistema cu√°ntico**, incluyendo las **probabilidades y fases cu√°nticas** de cada uno de los estados base.



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar compuerta X al qubit 1
qc = QuantumCircuit(2)
qc.x(1)
psi = Statevector(qc)

# Mostrar circuito con compuerta X
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)

````

### üßø ¬øQu√© estamos viendo en la QSphere?

La **QSphere** es una herramienta gr√°fica que nos permite observar el estado cu√°ntico completo del sistema, mostrando:

- Los **estados base** posibles (como `|00‚ü©`, `|01‚ü©`, `|10‚ü©`, `|11‚ü©`)
- Las **probabilidades** de cada uno (tama√±o del punto)
- Las **fases cu√°nticas** relativas (color del punto)

En esta visualizaci√≥n:

- El punto correspondiente a `|11‚ü©` es el √∫nico que aparece grande, lo que indica que el sistema est√° **completamente en ese estado**.
- Los dem√°s puntos (`|00‚ü©`, `|01‚ü©`, `|10‚ü©`) tienen **amplitud cero**, por lo que no aparecen visibles.
- El **color del punto** indica la **fase cu√°ntica** del estado. En este caso, al no haber superposici√≥n, no hay interferencia ni fases relativas importantes.

üìå La QSphere es especialmente √∫til para representar **estados con m√∫ltiples componentes y fases complejas**, como cuando hay **superposici√≥n o entrelazamiento**. Aqu√≠, como solo hay un estado (`|11‚ü©`), se ve un solo punto grande y bien definido.

# üìå Aplicar una compuerta X al qubit 0

Ahora aplicaremos la **compuerta X al qubit 0**, el menos significativo. Esto transforma el estado de `|00‚ü©` a `|10‚ü©`.


````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar compuerta X al qubit 0
qc = QuantumCircuit(2)
qc.x(0)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)

````

## ‚öôÔ∏è Secci√≥n 3 ‚Äì Aplicaci√≥n de compuertas individuales en ambos qubits

Hasta ahora hemos aplicado compuertas individuales a **un solo qubit a la vez**.  
En esta secci√≥n aplicaremos **una compuerta a cada qubit** dentro del mismo circuito.

Como ejemplo, aplicaremos la compuerta **X al qubit 0** y la compuerta **X al qubit 1**, lo cual transformar√° el estado inicial `|00‚ü©` en `|11‚ü©`.

Esto nos permitir√° observar c√≥mo se comporta el sistema cuando **ambos qubits cambian simult√°neamente**.

---




````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar compuertas X a ambos qubits
qc = QuantumCircuit(2)
qc.x(0)
qc.x(1)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar en notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)

````

## ‚öôÔ∏è Secci√≥n 4 ‚Äì Superposici√≥n en sistemas de 2 qubits

Hasta ahora hemos trabajado con qubits en estados cl√°sicos (`|0‚ü©`, `|1‚ü©`).  
En esta secci√≥n vamos a introducir el concepto de **superposici√≥n**, aplicando una **compuerta Hadamard** al **qubit 0** mientras que el qubit 1 permanece en `|0‚ü©`.

Esto genera un estado donde el sistema cu√°ntico est√° simult√°neamente en `|00‚ü©` y `|10‚ü©`.

Utilizaremos la **QSphere** para visualizar c√≥mo se distribuyen las amplitudes y las fases cu√°nticas en un sistema de 2 qubits en superposici√≥n.

---


````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito y aplicar Hadamard al qubit 0
qc = QuantumCircuit(2)
qc.h(0)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)
````

## ‚öôÔ∏è Secci√≥n 5 ‚Äì Entrelazamiento

Ahora aplicaremos una compuerta **CNOT** al sistema cu√°ntico que generamos previamente.  
Este sistema estaba en superposici√≥n: `(|00‚ü© + |10‚ü©)/‚àö2`, es decir, el **qubit 0** en superposici√≥n y el **qubit 1** en `|0‚ü©`.

La **CNOT** aplicar√° una compuerta X al qubit 1 **solo si el qubit 0 est√° en `|1‚ü©`**.

---

### üîó ¬øQu√© significa que est√©n entrelazados?

Este estado es **entrelazado**, lo que significa que los qubits est√°n conectados:

- Si mides el qubit 0 y da `0`, el qubit 1 tambi√©n ser√° `0`
- Si mides el qubit 0 y da `1`, el qubit 1 tambi√©n ser√° `1`

Aunque no los mires al mismo tiempo, **los resultados siempre estar√°n correlacionados**.

Este tipo de conexi√≥n cu√°ntica es lo que hace al entrelazamiento tan especial y √∫til en la computaci√≥n cu√°ntica.

---

En Qiskit, esta compuerta se codifica como:

```python
qc.cx(control, target)



````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Crear circuito: Hadamard en qubit 0, luego CNOT (0 controla a 1)
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Estado vectorial del sistema:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø Visualizaci√≥n en QSphere:")
plot_state_qsphere(psi)
````

## ‚öôÔ∏è Secci√≥n 6 ‚Äì Aplicar m√∫ltiples compuertas en un mismo qubit

Hasta ahora solo hemos aplicado **una compuerta por qubit**.  
En esta secci√≥n aprenderemos que es posible aplicar **varias compuertas en el mismo qubit, una tras otra**.

Esto es muy com√∫n en circuitos cu√°nticos, y permite **modificar gradualmente el estado de un qubit**.

Veremos dos casos:

- Aplicar `H` seguido de `X` al qubit 0
- Aplicar `X` seguido de `H` al mismo qubit

Veremos que **el orden importa** y el estado final cambia dependiendo del orden.

---


````python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_state_qsphere
from IPython.display import display

# Caso 1: aplicar Hadamard luego X al qubit 0
qc = QuantumCircuit(1)
qc.x(0)
qc.h(0)
psi = Statevector(qc)

# Mostrar circuito
display(qc.draw('mpl'))

# Mostrar vector de estado
print("üß™ Vector de estado:")
display(array_to_latex(psi.data))

# Mostrar notaci√≥n ket
print("üß™ Estado en notaci√≥n ket:")
display(psi.draw('latex'))

# Mostrar QSphere
print("üßø QSphere del estado:")
plot_state_qsphere(psi)
````
